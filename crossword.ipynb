```python
import numpy as np

# Define the grid dimensions and the grid itself ('.' for white, '#' for black)
grid_height = 10
grid_width = 10
grid = np.full((grid_height, grid_width), '.')

# Define some black squares (blocked cells)
blocked_cells = [(1, 1), (2, 2), (3, 3), (4, 4)]
for cell in blocked_cells:
    grid[cell] = '#'

# Define a list of words to place in the grid
word_list = ["python", "code", "grid", "words", "algorithm", "place"]

def can_place_word(grid, word, row, col, direction):
    """ Check if a word can be placed at the given position in the specified direction. """
    if direction == 'horizontal':
        if col + len(word) > grid.shape[1]:  # Word goes out of grid
            return False
        for i in range(len(word)):
            if grid[row, col + i] == '#' or (grid[row, col + i] != '.' and grid[row, col + i] != word[i]):
                return False
    elif direction == 'vertical':
        if row + len(word) > grid.shape[0]:  # Word goes out of grid
            return False
        for i in range(len(word)):
            if grid[row + i, col] == '#' or (grid[row + i, col] != '.' and grid[row + i, col] != word[i]):
                return False
    return True

def place_word(grid, word, row, col, direction):
    """ Place a word in the grid at the given position in the specified direction. """
    if direction == 'horizontal':
        for i in range(len(word)):
            grid[row, col + i] = word[i]
    elif direction == 'vertical':
        for i in range(len(word)):
            grid[row + i, col] = word[i]

def print_grid(grid):
    """ Print the grid. """
    for row in grid:
        print("".join(row))

# Try to place each word in the grid
for word in word_list:
    placed = False
    for row in range(grid_height):
        for col in range(grid_width):
            if can_place_word(grid, word, row, col, 'horizontal'):
                place_word(grid, word, row, col, 'horizontal')
                placed = True
                break
            if can_place_word(grid, word, row, col, 'vertical'):
                place_word(grid, word, row, col, 'vertical')
                placed = True
                break
        if placed:
            break

# Print
print_grid(grid)

```

```python
import numpy as np

# Define the grid dimensions and the grid itself ('.' for white, '#' for black)
grid_height = 10
grid_width = 10
grid = np.full((grid_height, grid_width), '.')

# Define some black squares (blocked cells)
blocked_cells = [(1, 1), (2, 2), (3, 3), (4, 4)]
for cell in blocked_cells:
    grid[cell] = '#'

# Define a list of words to place in the grid
word_list = ["python", "code", "grid", "words", "algorithm", "place"]

def can_place_word(grid, word, row, col, direction):
    """ Check if a word can be placed at the given position in the specified direction. """
    if direction == 'horizontal':
        if col + len(word) > grid.shape[1]:  # Word goes out of grid
            return False
        for i in range(len(word)):
            if grid[row, col + i] == '#' or (grid[row, col + i] != '.' and grid[row, col + i] != word[i]):
                return False
    elif direction == 'vertical':
        if row + len(word) > grid.shape[0]:  # Word goes out of grid
            return False
        for i in range(len(word)):
            if grid[row + i, col] == '#' or (grid[row + i, col] != '.' and grid[row + i, col] != word[i]):
                return False
    return True

def place_word(grid, word, row, col, direction):
    """ Place a word in the grid at the given position in the specified direction. """
    if direction == 'horizontal':
        for i in range(len(word)):
            grid[row, col + i] = word[i]
    elif direction == 'vertical':
        for i in range(len(word)):
            grid[row + i, col] = word[i]

def find_place_for_word(grid, word):
    """ Find the best place for a word with maximum overlap. """
    best_score = -1
    best_position = None
    for row in range(grid_height):
        for col in range(grid_width):
            for direction in ['horizontal', 'vertical']:
                if can_place_word(grid, word, row, col, direction):
                    score = 0
                    if direction == 'horizontal':
                        for i in range(len(word)):
                            if grid[row, col + i] == word[i]:
                                score += 1
                    elif direction == 'vertical':
                        for i in range(len(word)):
                            if grid[row + i, col] == word[i]:
                                score += 1
                    if score > best_score:
                        best_score = score
                        best_position = (row, col, direction)
    return best_position

def print_grid(grid):
    """ Print the grid. """
    for row in grid:
        print("".join(row))

# Try to place each word in the grid
for word in word_list:
    position = find_place_for_word(grid, word)
    if position:
        row, col, direction = position
        place_word(grid, word, row, col, direction)

# Print the resulting grid
print_grid(grid)
```
