```python
def get_horizontal_word(grid, row, col):
    word = ''
    start_col = col
    while start_col >= 0 and grid[row, start_col] != '.':
        start_col -= 1
    start_col += 1
    while start_col < len(grid[0]) and grid[row, start_col] != '.':
        word += grid[row, start_col]
        start_col += 1
    return word

def get_vertical_word(grid, row, col):
    word = ''
    start_row = row
    while start_row >= 0 and grid[start_row, col] != '.':
        start_row -= 1
    start_row += 1
    while start_row < len(grid) and grid[start_row, col] != '.':
        word += grid[start_row, col]
        start_row += 1
    return word
```

```python
from spellchecker import SpellChecker

def check_spelling(word):
    spell = SpellChecker()
    if word in spell:
        return True
    else:
        return False
```

```python
def validate_grid(grid, valid_words):
    for row in range(len(grid)):
        for col in range(len(grid[0])):
            if grid[row, col] != '.':
                if col == 0 or grid[row, col - 1] == '.':
                    horizontal_word = get_horizontal_word(grid, row, col)
                    if len(horizontal_word) > 1 and not check_spelling(horizontal_word):
                        return False
                if row == 0 or grid[row - 1, col] == '.':
                    vertical_word = get_vertical_word(grid, row, col)
                    if len(vertical_word) > 1 and not check_spelling(vertical_word):
                        return False
    return True

def add_word(grid, word, row, col, direction, valid_words):
    temp_grid = grid.copy()
    if direction == 'horizontal':
        if col + len(word) > len(grid[0]):
            return False
        for i, char in enumerate(word):
            if temp_grid[row, col + i] != '.' and temp_grid[row, col + i] != char:
                return False
            temp_grid[row, col + i] = char
    elif direction == 'vertical':
        if row + len(word) > len(grid):
            return False
        for i, char in enumerate(word):
            if temp_grid[row + i, col] != '.' and temp_grid[row + i, col] != char:
                return False
            temp_grid[row + i, col] = char

    if validate_grid(temp_grid, valid_words):
        grid[:] = temp_grid
        return True
    return False
```
